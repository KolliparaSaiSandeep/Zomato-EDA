```python
country_data = {'Country Code': [162, 30, 216, 14, 37, 184, 214, 1, 94, 148, 215, 166, 189, 191, 208],
                'Country': ['Country_A', 'Country_B', 'Country_C', 'Country_D', 'Country_E',
                            'Country_F', 'Country_G', 'Country_H', 'Country_I', 'Country_J',
                            'Country_K', 'Country_L', 'Country_M', 'Country_N', 'Country_O']}

df_country = pd.DataFrame(country_data)
df_country
```

```python
final_df=pd.merge(df,df_country,on='Country Code', how='left')
final_df
```

```python
##To check Data Types
final_df.dtypes
```

```python
country_names=final_df.Country.value_counts().index
country_names
```

```python
country_val=final_df.Country.value_counts().values
country_val
```

```python
plt.pie(country_val[:3],labels=country_names[:3])
```

Observation:Zomato maximum records are from Country_H,after that Country_C and Country_K

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size()
ratings
```

```python
ratings=final_df.groupby(["Aggregate rating","Rating color","Rating text"]).size().reset_index().rename(columns={0:"Rating Count"})
```

**Observation**

When Rating is between 4.5 to 4.9---> Excellent

When Rating are between 4.0 to 3.4--->very good

when Rating is between 3.5 to 3.9----> good

when Rating is between 3.0 to 3.4----> average

when Rating is between 2.5 to 2.9----> average

when Rating is between 2.0 to 2.4----> Poor  

```python
ratings.head()
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['white','red','orange','yellow','green','green'])
```

Observation:

Not Rated count is very high
Maximum number of rating are between 2.5 to 3.4

```python
## Count plot
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
##for clarity btw value_counts and size

import pandas as pd

# Create a sample DataFrame
data = {'Category': ['A', 'B', 'A', 'C', 'B', 'A', 'C']}
df = pd.DataFrame(data)

# Using value_counts() to get the count of unique values in the 'Category' column
value_counts_result = df['Category'].value_counts()
print("Using value_counts():")
print(value_counts_result)

# Using size to get the total number of elements in the 'Category' column
size_result = df['Category'].size
print("\nUsing size:")
print(size_result)
```
